```python
import numpy as np
import matplotlib.pyplot as plt
from vamas import Vamas

# Initialize VAMAS file parameters
file_path = "C:/Users/mdram/Computational_Projects/BAM reference data_ XPS raw data of Al-coated titania nanoparticles/data/jrc620001-xps.vms"
vms_data = Vamas(file_path)
hv = 1486.69  # Al K-alpha source (eV)
energy_shift = -4.2  # Default calibration shift calculated from C 1s peak position

def get_calibrated_be_and_y(block_index):
    """Extracts raw intensity values and returns calibrated Binding Energy (eV)"""
    block = vms_data.blocks[block_index]
    y = np.array(block.corresponding_variables[0].y_values)
    ke = np.array([block.x_start + (i * block.x_step) for i in range(len(y))])
    be = (hv - ke) + energy_shift  # Apply source energy and calibration shift
    return be, y
```

```python
# Extract C 1s (Block 6 is at index 5)
be_c1s, y_c1s = get_calibrated_be_and_y(5)
measured_peak = be_c1s[np.argmax(y_c1s)]

plt.figure(figsize=(8, 4.5))
plt.plot(be_c1s, y_c1s, color="black", label="Experimental Data", linewidth=1.5)
plt.axvline(284.8, color="red", linestyle="--", label="Calibration Reference (284.8 eV)")
plt.axvline(measured_peak, color="blue", linestyle=":", label=f"Measured Peak ({measured_peak:.1f} eV)")

plt.gca().invert_xaxis()
plt.title("Cell 1: C 1s Calibration & Charging Assessment", fontweight="bold")
plt.xlabel("Binding Energy (eV)")
plt.ylabel("Intensity (Counts)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
analyzed_regions = {
    2: {"name": "F 1s",  "rsf": 1.000},
    3: {"name": "O 1s",  "rsf": 0.711},
    4: {"name": "Ti 2p", "rsf": 2.001},
    5: {"name": "K 2p",  "rsf": 1.466},
    13:{"name": "Al 2p", "rsf": 0.193}
}

quant_results = []
for bid, info in analyzed_regions.items():
    be, y = get_calibrated_be_and_y(bid - 1)
    bg = np.linspace(y[0], y[-1], len(y))
    y_corrected = np.maximum(y - bg, 0)
    raw_area = np.trapz(y_corrected, be)
    quant_results.append({"Element": info["name"], "Rel_I": raw_area / info["rsf"]})

# Compute Atomic percentages
quant_df = pd.DataFrame(quant_results)
quant_df["Atomic %"] = (quant_df["Rel_I"] / quant_df["Rel_I"].sum()) * 100

# Plotting the results
plt.figure(figsize=(7, 4.5))
plt.bar(quant_df["Element"], quant_df["Atomic %"], color=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"])
for idx, row in quant_df.iterrows():
    plt.text(idx, row["Atomic %"] + 1, f"{row['Atomic %']:.1f}%", ha='center', va='bottom', fontweight='bold')

plt.title("Cell 2: Surface Elemental Concentration (at %)", fontweight="bold")
plt.ylabel("Atomic Concentration (%)")
plt.ylim(0, max(quant_df["Atomic %"]) * 1.15)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
from scipy.optimize import curve_fit

def gaussian(x, amp, mean, sigma):
    return amp * np.exp(-((x - mean) ** 2) / (2 * sigma ** 2))

def dual_peak_model(x, a1, m1, s1, a2, m2, s2):
    return gaussian(x, a1, m1, s1) + gaussian(x, a2, m2, s2)

# Extract O 1s (Block 3 is at index 2)
be_o1s, y_o1s = get_calibrated_be_and_y(2)
y_sub = y_o1s - np.linspace(y_o1s[0], y_o1s[-1], len(y_o1s))

# Curve fit
initial_guess = [max(y_sub), 530.1, 1.0, max(y_sub)*0.3, 532.2, 1.2]
popt, _ = curve_fit(dual_peak_model, be_o1s, y_sub, p0=initial_guess)

# Generate fit curves
fit_total = dual_peak_model(be_o1s, *popt)
fit_peak1 = gaussian(be_o1s, *popt[0:3])
fit_peak2 = gaussian(be_o1s, *popt[3:6])

plt.figure(figsize=(8, 4.5))
plt.plot(be_o1s, y_sub, 'k.', label="Experimental Data (BG Subtracted)", alpha=0.5)
plt.plot(be_o1s, fit_total, 'r-', label="Overall Fitted Envelope", linewidth=2)
plt.fill_between(be_o1s, fit_peak1, color='blue', alpha=0.2, label=f"Metal-Oxide (Ti-O/Al-O) @ {popt[1]:.1f} eV")
plt.fill_between(be_o1s, fit_peak2, color='green', alpha=0.2, label=f"Hydroxyls/Surface-O @ {popt[4]:.1f} eV")

plt.gca().invert_xaxis()
plt.title("Cell 3: Chemical States of Oxygen (O 1s Deconvolution)", fontweight="bold")
plt.xlabel("Binding Energy (eV)")
plt.ylabel("Intensity (Net Counts)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# CELL 2: SURFACE ELEMENTAL QUANTIFICATION
# ==========================================
import scipy.integrate as integrate

# Element Scopes (Block IDs & corresponding RSF values under Al K-alpha)
# Note: RSF values are Scofield library standard estimates
analyzed_regions = {
    2: {"name": "F 1s",  "rsf": 1.000},
    3: {"name": "O 1s",  "rsf": 0.711},
    4: {"name": "Ti 2p", "rsf": 2.001},
    5: {"name": "K 2p",  "rsf": 1.466},
    13:{"name": "Al 2p", "rsf": 0.193}
}

quant_results = []

for bid, info in analyzed_regions.items():
    block = vms_data.blocks[bid - 1]
    y = np.array(block.corresponding_variables[0].y_values)
    x = hv - np.array([block.x_start + (i * block.x_step) for i in range(len(y))])
    
    # 1. Simple Linear Background Subtraction
    bg = np.linspace(y[0], y[-1], len(y))
    y_corrected = np.maximum(y - bg, 0)
    
    # 2. Integrate area under peak
    raw_area = np.trapz(y_corrected, x)
    
    # 3. Calculate relative intensity (Area / RSF)
    rel_intensity = raw_area / info["rsf"]
    
    quant_results.append({
        "Element": info["name"],
        "Raw Area": raw_area,
        "RSF": info["rsf"],
        "Rel Intensity": rel_intensity
    })

# Compute final Atomic Percentages
quant_df = pd.DataFrame(quant_results)
total_rel_intensity = quant_df["Rel Intensity"].sum()
quant_df["Atomic %"] = (quant_df["Rel Intensity"] / total_rel_intensity) * 100

print(f"--- 2. SURFACE ATOMIC CONCENTRATIONS ---")
print(quant_df[["Element", "Raw Area", "RSF", "Atomic %"]].to_string(index=False))
```

```python
def shirley_bg(be, y, tol=1e-6, maxit=50):
    """Iterative Shirley background: assumes be is ordered high->low BE."""
    y = np.asarray(y, dtype=float)
    n = len(y)
    B = np.linspace(y[-1], y[0], n)          # linear seed between endpoints
    for _ in range(maxit):
        total = np.trapezoid(y - B, x=np.arange(n))
        if total == 0:
            break
        Bnew = np.empty(n)
        for i in range(n):
            above = np.trapezoid((y - B)[i:], x=np.arange(i, n))
            Bnew[i] = y[-1] + (y[0] - y[-1]) * above / total
        if np.max(np.abs(Bnew - B)) < tol:
            B = Bnew
            break
        B = Bnew
    return B

results = []
for region in df.region.unique():
    if region == 'widescan':
        continue
    sub = df[df.region == region].sort_values('BE_eV', ascending=False)
    be, y = sub.BE_eV.values, sub.counts.values
    bg = shirley_bg(be, y)
    y_net = np.clip(y - bg, 0, None)

    area = abs(np.trapezoid(y_net, x=be))
    peak_be = be[np.argmax(y_net)]
    centroid = np.sum(be * y_net) / np.sum(y_net)

    results.append(dict(region=region, peak_BE=peak_be,
                         centroid_BE=centroid, shirley_area=area))

peaks = pd.DataFrame(results).sort_values('peak_BE', ascending=False)
peaks.to_csv('peak_results.csv', index=False)
peaks
```
